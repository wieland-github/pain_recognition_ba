# Code für feature Extraction wird hier ausgeführt

```python
from pathlib import Path
import numpy as np
import pandas as pd
from pycatch22 import catch22_all
from scripts.feature_extraction_methods import extract_catch22_features, time_domain_features, extract_all_features_from_one_signal, extract_features_dataset
from scripts.myml import make_signal_groups
```

### Globale Paths und Konstanten

```python
DATASET_PATH = Path("dataset/np-dataset")
X_PATH = DATASET_PATH / "X.npy"
OUT_CATCH22_FEATURES = DATASET_PATH / "X_catch22.npy"
OUT_CATCH22_FEATURE_NAMES = DATASET_PATH / "feature_names.npy"

SENSOR_NAMES  = ["Bvp", "Eda_E4", "Resp", "Eda_RB", "Ecg", "Emg"]
SAMPLING_RATE = 250
```

### Daten Laden

```python
X = np.load(X_PATH)          # (2495, 2500, 6, 1)
X = X[..., 0]                # (2495, 2500, 6)  
n_samples, n_timepoints, n_sensors = X.shape
print(f"  X shape after dropping channel axis: {X.shape}")
print (f"  Number of samples: {n_samples}")
print (f"  Number of time points: {n_timepoints}")
print (f"  Number of sensors: {n_sensors}")
```

### Feature Extraction mit einem sample testen

```python
vals, names = extract_all_features_from_one_signal(X[0, :, 0])

print(f"Number of features extracted from one signal: {len(vals)}")
print(f"Feature names: {names}")
```

### Feature Extraction für alle samples

```python
X_features, feature_names = extract_features_dataset(X, SENSOR_NAMES)
print(f"Extracted features shape: {X_features.shape}") 
print(f"Anzahl Features: {len(feature_names)}")
```

### Features als numpy matrix speichern

```python
np.save(OUT_CATCH22_FEATURES, X_features)
np.save(OUT_CATCH22_FEATURE_NAMES, feature_names)
print(f"Features und Namen gespeichert unter: {OUT_CATCH22_FEATURES} und {OUT_CATCH22_FEATURE_NAMES}")
print(f"Feature matrix shape = {X_features.shape}")
print(f"Column names len = {len(feature_names)}")
```
